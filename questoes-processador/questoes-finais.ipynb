## QUESTÃO 1:



A UCP (Unidade Central de Processamento), também chamada de CPU, é responsável por coordenar, decodificar e executar as instruções de um programa. Ela realiza operações aritméticas, lógicas e de desvio, além de controlar o funcionamento dos demais componentes do computador.

Seus principais componentes são:

- CI ou PC (Contador de Instruções / Program Counter): armazena o endereço da próxima instrução que será executada.
- RI (Registrador de Instruções): armazena a instrução atual para que ela seja decodificada e executada.
- ACC (Acumulador): registrador utilizado para armazenar resultados temporários das operações realizadas pela ULA.
- ULA (Unidade Lógica e Aritmética): executa operações aritméticas (como soma e subtração) e operações lógicas (como AND, OR e NOT).
- Unidade de Controle (UC): coordena a execução das instruções, controlando o fluxo de dados entre os registradores, a memória e a ULA.

## QUESTÃO 2:



A ULA (Unidade Lógica e Aritmética) é o componente da CPU responsável por realizar todas as operações de cálculo e comparação exigidas pelas instruções de um programa. Suas principais funções são:

- Operações aritméticas: soma, subtração, multiplicação, divisão e incremento/decremento de valores.
- Operações lógicas: AND, OR, NOT, XOR e demais operações bit a bit.
- Operações de comparação: verificar se um valor é maior, menor ou igual a outro, usadas para decisões de desvio condicional.
- Deslocamento de bits (shift): deslocar os bits de um valor para a esquerda ou para a direita.

Os operandos utilizados pela ULA vêm dos registradores (como o ACC) ou da memória, e o resultado da operação é armazenado novamente em um registrador, geralmente o ACC, para ser usado pelas instruções seguintes.

## QUESTÃO 3:



O ACC (Acumulador) é um registrador da CPU utilizado para armazenar temporariamente os resultados das operações realizadas pela ULA (Unidade Lógica e Aritmética).

Ele serve como um local de trabalho para o processador, permitindo que resultados intermediários sejam guardados antes de serem enviados para a memória ou utilizados em novas operações. Em muitos processadores, o tamanho do ACC também define o tamanho da palavra de processamento da arquitetura.

## QUESTÃO 5:



A Unidade de Controle (UC) é o componente da CPU responsável por comandar e sincronizar todas as ações necessárias à execução das instruções de um programa. Suas principais funções são:

- Buscar a instrução na memória, utilizando o endereço indicado pelo CI, e transferi-la para o RI.
- Decodificar a instrução, identificando qual operação deve ser realizada e quais operandos serão utilizados.
- Gerar os sinais de controle que comandam os demais componentes (registradores, ULA e memória), determinando quando e como cada um deve atuar.
- Coordenar o fluxo de dados entre a memória, os registradores e a ULA durante a execução da instrução.
- Controlar a sequência de execução do programa, atualizando o CI para a próxima instrução ou desviando a execução quando necessário.

## QUESTÃO 6:



Na prática, o CI e o REM normalmente possuem o mesmo número de bits, pois ambos armazenam endereços de memória. Se fossem diferentes, o CI nunca poderia ser menor que o REM, pois isso impediria o acesso a parte da memória. Portanto, o mais comum é que tenham o mesmo tamanho.

## QUESTÃO 7:



Como as instruções não possuem tamanho fixo, o CI não pode ser incrementado de um valor constante após a busca de cada instrução. A solução prática é fazer o incremento em duas etapas:

1. Primeiro, a UCP busca apenas o byte referente ao código de operação (8 bits, ou seja, 1 célula de memória), incrementando o CI em 1 nesse momento.
2. Ao decodificar o código de operação, a Unidade de Controle já identifica a qual dos três tipos a instrução pertence (A, B ou C) e, portanto, sabe quantos bytes adicionais compõem o restante da instrução (1 byte a mais para o tipo A, totalizando 2 bytes; 3 bytes a mais para o tipo B, totalizando 4 bytes; ou 5 bytes a mais para o tipo C, totalizando 6 bytes).
3. Com essa informação, a UCP busca os bytes restantes da instrução e incrementa o CI novamente pela quantidade correspondente, de modo que, ao final da busca, o CI aponte exatamente para o primeiro byte da instrução seguinte.

Ou seja, o incremento automático do CI passa a ser variável, dependendo do tipo de instrução identificado pelo código de operação, em vez de um valor fixo.

## QUESTÃO 8:



Sem a etapa de incremento automático do CI, o processador não tem como saber, por conta própria, o endereço da próxima instrução a ser executada. Uma alternativa prática é fazer com que cada instrução do programa carregue explicitamente, como parte de seu próprio código, o endereço da instrução seguinte que deve ser executada.

Nesse método, toda instrução passaria a ter um campo adicional contendo o endereço da próxima instrução. Ao final da execução de cada instrução, em vez de incrementar o CI, a Unidade de Controle simplesmente copiaria esse endereço para o CI. Dessa forma, o programa funcionaria como uma espécie de lista encadeada de instruções, na qual cada instrução aponta para a próxima, garantindo a continuidade da execução mesmo sem o incremento automático.

## QUESTÃO 9:



O registrador é o CI (Contador de Instruções), também chamado de PC (Program Counter). Ele armazena o endereço da próxima instrução a ser buscada na memória e, por isso, é o registrador cujo conteúdo determina a ordem em que as instruções de um programa são processadas.

O CI está localizado dentro da CPU, mais especificamente associado à Unidade de Controle, que o utiliza para buscar cada instrução na memória e atualizá-lo (por incremento ou por desvio) para indicar a próxima instrução a ser executada.

## QUESTÃO 11:



O registrador é o REM (Registrador de Endereço de Memória), também chamado de MAR (Memory Address Register).

Isso ocorre porque o REM é o registrador utilizado para armazenar o endereço da célula de memória que será acessada em uma operação de leitura ou escrita. A quantidade de bits do REM define quantos endereços diferentes podem ser representados: com n bits, é possível representar 2^n endereços distintos. Logo, a capacidade máxima de memória que o computador pode endereçar é igual a 2^n células, sendo n o número de bits do REM. Por isso, o tamanho do REM impõe um limite direto à capacidade de memória do computador.

## QUESTÃO 15:



Quando se diz que um microcomputador é "de 8 bits" ou "de 16 bits", normalmente a referência é ao tamanho da palavra do processador, e não ao tamanho da célula de memória.

A célula de memória (MP) tem seu tamanho definido por convenção de endereçamento, sendo, na grande maioria das arquiteturas, fixada em 8 bits (1 byte), independentemente do processador utilizado. Já a palavra é definida pela largura dos registradores internos da CPU (como o ACC), da ULA e do barramento de dados, ou seja, pela quantidade de bits que o processador consegue manipular e processar de uma só vez em uma operação.

Portanto, a base desses dois conceitos é diferente: o tamanho da célula é uma convenção de organização da memória (o menor bloco endereçável), enquanto o tamanho da palavra é uma característica do processador (a quantidade de bits processada por operação). Um microcomputador "de 16 bits" possui palavra de 16 bits, mesmo que sua memória continue organizada em células de 8 bits.

## QUESTÃO 18:



No processamento sequencial, cada instrução só começa a ser buscada, decodificada e executada depois que a instrução anterior tiver completado todas as suas etapas. Ou seja, as etapas do ciclo de instrução (busca, decodificação e execução) ocorrem uma de cada vez, e o processador fica ocioso em relação às demais etapas enquanto uma instrução está sendo processada.

No processamento em pipeline, as etapas do ciclo de instrução são executadas em paralelo para instruções diferentes, de forma semelhante a uma linha de montagem. Enquanto uma instrução está sendo executada, a próxima já pode estar sendo decodificada, e uma terceira já pode estar sendo buscada na memória, tudo ao mesmo tempo. Isso faz com que, embora cada instrução individualmente ainda leve o mesmo tempo para ser concluída, o processador consiga iniciar uma nova instrução a cada ciclo, aumentando significativamente o desempenho geral, medido pela taxa de instruções concluídas por unidade de tempo.

## QUESTÃO 22:



a) Cada instrução possui um único formato (C.Op | Reg | Op.2), e o RI, que armazena a instrução completa, tem 32 bits (RI = 2C4F08D9, ou seja, 8 dígitos hexadecimais). Como todos os endereços possuem 16 bits, o campo Op.2 precisa ter 16 bits para poder representar um endereço completo de memória. Já o campo Reg precisa de 4 bits, pois o processador possui 16 registradores de dados. Assim, a largura do campo C.Op é obtida por:

C.Op = 32 - Reg - Op.2 = 32 - 4 - 16 = 12 bits

b) Como o processador possui 16 registradores de dados, e 2^4 = 16, são necessários 4 bits para o campo Reg, de modo a permitir que qualquer um dos 16 registradores seja referenciado por essa instrução.

c) É preciso observar a notação usada na tabela: nas instruções MOV e MOVM, o campo Op aparece entre parênteses - "(Op.)" -, o que indica um acesso indireto à memória, ou seja, o conteúdo da célula de endereço Op. Já nas instruções JMP, JP, ADD e SUB, o campo Op aparece sem parênteses - apenas "Op." -, o que indica que esse valor é usado diretamente (endereçamento imediato), pois ele já está contido na própria instrução, sem necessidade de uma busca adicional à memória.

Para a instrução de C.Op igual a 11 (JP R, Op - "Se R > 0, então: CI ← Op."), a UCP testa o conteúdo do registrador R e, se a condição for satisfeita, carrega diretamente o valor de Op (já presente na instrução) no CI. Como não há acesso indireto à memória, é necessário apenas 1 acesso à memória: a busca da instrução completa (C.Op, Reg e Op.2 já vêm juntos nessa mesma leitura).

Para a instrução de C.Op igual a 21 (SUB R, Op - "(R) ← (R) - Op."), o valor de Op também é usado diretamente, como operando imediato, e o resultado da subtração é gravado apenas no registrador R. Assim, também é necessário apenas 1 acesso à memória: a busca da instrução.

Ou seja, tanto para C.Op = 11 quanto para C.Op = 21, basta 1 acesso à memória, pois nenhuma das duas instruções realiza leitura ou escrita indireta em memória - diferentemente de instruções como MOV (01) e MOVM (02), que precisariam de 2 acessos (1 para buscar a instrução e 1 para ler ou escrever o operando na célula de memória apontada por Op).
